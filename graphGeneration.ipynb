```python
import networkx as nx
import datetime
import pandas as pd
from dateutil.parser import parse
import numpy
import pickle
```

```python
data = pd.read_csv("Combined.csv")
df= pd.DataFrame(data)
G=nx.Graph()
```

```python
#establishing start date and end date for the dataset 
start_date=parse(df["Created"][0])
end_date=parse("2/01/2040")

#the increment value
delta=datetime.timedelta(days=30)
print(start_date)
print (end_date)
```

```python
while end_date.date()>start_date.date():
    
    
    #There are 457 rows in the dataset
        for i in range(0,453):
            
            #add nodes
            
            #domain
            G.add_node(df["Domain"][i], Category=df["Category"][i],Type="Domain")
           #registration address
            G.add_node(df["RegistrationAddress"][i],Category=df["Category"][i],Type="Address")
            
            #there are multiple server entries in one cell so we split it first
            server=str(df["NameServers"][i])
            if(server=="nan"):
                pass
            else:
                servers_list=server.split(",")
                for item in servers_list:
                    G.add_node(item,Category=df["Category"][i],Type="Server")
            #phone numbers
            phone=str(df["Phone"][i])
            if(phone=="nan"):
                pass
            else:
                phone_list=phone.split(",")
                for item in phone_list:
                    G.add_node(item,Category=df["Category"][i],Type="phone")


            
            #email
            emails=str(df["Email"][i])
            
            if(emails=="nan"):
                pass
            else:
                emails_list=emails.split(",")
               
                for item in emails_list:
                    G.add_node(item,Category=df["Category"][i],Type="Email")
                    
                    
            #for edges

            created=parse(df["Created"][i])
         
            expiry=parse(df["Expiry"][i])


            if(start_date.date()>created.date() and start_date.date()<expiry.date()):
                
                #domain -- Registration Address
                G.add_edge(df["Domain"][i],df["RegistrationAddress"][i])
                
                #domain -- Name servers
                server=str(df["NameServers"][i])
                if(server=="nan"):
                   
                    pass
                else:
                    
                    servers_list=server.split(",")
                for item in servers_list:
                    G.add_edge(item,df["Domain"][i])
                
                #domain -- Phone numbers
                phone_num=str(df["Phone"][i])
                if(phone_num=="nan"):
                    
                    pass
                else:
                    
                    phone_list=phone_num.split(",")
                for item in phone_list:
                    G.add_edge(item,df["Domain"][i])
                    
                #domain -- email
                emails=str(df["Email"][i])   
                if(emails=="nan"):
                    
                    pass
                else:
                    
                    emails_list=emails.split(",")
               
                for item in emails_list:
                    G.add_edge(item,df["Domain"][i])
            
       
        
        nx.write_gpickle(G,str(start_date.date())+".pickle")
        G.clear()
        start_date+=delta
        
        
        
            
            
        
```
